# Missing Value Imputation methods

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%%time
df=pd.read_csv(r"F:\titanic_train.csv",sep=',')
df.head()
```



### 1. Mean/median imputation

```python
%%time
def imputation(df,variable):
    mean=df[variable].mean()
    median=df[variable].median()
    df[variable+"_mean"]=df[variable].fillna(mean)
    df[variable+"_median"]=df[variable].fillna(median)
    return df.head()
```

```python
imputation(df,"age")
```

### 2. Random variable selection imputation

```python
%%time
def random(df,variable):
    df[variable +'_random']=df[variable]
    random_sample=df[variable].dropna().sample(df[variable].isnull().sum(),random_state=0)
    random_sample.index=df[df[variable].isnull()].index
    df.loc[df[variable].isnull(),variable+'_random']=random_sample
    return df.head()
   
```

```python
random(df,"age")
```

```python
df.head()
```

### 3. Extreme Value Imputation

```python
%%time
def extreme(df,variable):
    extreme=df[variable].mean()+3*df[variable].std()
    df[variable +'_extreme']=df[variable].fillna(extreme)
    return df.head()
```

```python
extreme(df,"age")
```

### 4. Arbitrary value imputation

```python
%%time
def Arbitrary(df,variable):
    df[variable+'_zero']=df[variable].fillna(0)
    df[variable+'_hundred']=df[variable].fillna(100)
    
```

```python
Arbitrary(df,"age")
```

```python
df.head()
```

### 5.Replacing the missing values with one's and zero's

```python
%%time
def nan(df,variable):
    df[variable+'_new']=np.where(df[variable].isnull(),1,0)
    
```

```python
nan(df,"age")
```
